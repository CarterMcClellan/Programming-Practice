```python
def collatz(n):
    while n != 1:
        if n % 2 == 0:
            yield int(n)
            n /= 2
        else:
            yield int(n)
            n = 3*n + 1
    yield 1
```

```python
for i in collatz(3289074298056378256328907456324872451):
    if i % 2 == 0:
        print("Even", i)
    else:
        print("Odd", i)
```
