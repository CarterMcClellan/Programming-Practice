```python
from math import factorial
```

```python
def choose(n, k):
    return factorial(n) // (factorial(k)*factorial(n - k))
```

```python
# paths from (0, 0) to (2, 2) are length 4
# there are only 2 direction (r, d)
# possible sequences are therefore

# (r, d, r, d), (d, r, d, r)
# (r, r, d, d), (d, d, r, r)
# (r, d, d, r), (d, r, r, d)

# how many different ways are there to combine 2 rights and 2 downs

# 4 possible posisition, choose 2 possible positions for r, d filling the other positions
choose(4, 2) # we see this matches the set which we generated by hand
```

```python
# from this we know the number of paths for an N dimensional lattice to be
# choose(2N, N)
# remember 2N because we will need to make both 20 lefts and 20 downs to reach the end
choose(40, 20)
```
